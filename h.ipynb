```python
import numpy as np
import cv2 as cv

# Create empty image
img = np.zeros([750, 750, 3], dtype=np.uint8)
image = cv.imread('sample.jpg')

x1, y1 = 500, 500
x2, y2 = 250, 700

cv.line(img, (x1, y1), (x2, y2), (255, 255, 255), 5)

# Define the kernel
line_length = max(abs(x2 - x1), abs(y2 - y1))
kernel_size = line_length + 5
kernel = np.ones((kernel_size, kernel_size), dtype=np.float32)
kernel /= kernel_size ** 2

# Apply the motion blur effect to the line using the convolution operation
blurred_line = cv.filter2D(img, -1, kernel)

# Combine the blurred line with the original image
output_img = cv.add(image, blurred_line)

# Display the original image, line, and the output with motion blur
cv.imshow('Original Image', image)
cv.imshow('Line with Motion Blur', blurred_line)
cv.imshow('Output', output_img)

# Wait for a key to be pressed
cv.waitKey(0)

# Clean up
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv

# Load the image
image = cv.imread('Circle Objects.png')

# Convert the image to grayscale
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred = cv.GaussianBlur(gray, (5, 5), 0)

# Apply HoughCircles to detect circles in the image
circles = cv.HoughCircles(blurred, cv.HOUGH_GRADIENT, 1, 20, param1=50, param2=30, minRadius=0, maxRadius=0)

if circles is not None:
    # Convert the coordinates and radius to integers
    circles = np.round(circles[0, :]).astype("int")

    # Iterate over detected circles
    for (x, y, r) in circles:
        # Draw a white dot in the center of the circle
        cv.circle(image, (x, y), 2, (255, 255, 255), -1)
        # Draw the circle outline
        cv.circle(image, (x, y), r, (0, 255, 0), 2)

# Display the image with the circle center marked
cv.imshow('Circle with Center', image)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

def draw_circle_on_non_white_pixels(image_path):
    # Load the image
    image = cv.imread(image_path)

    # Create a mask of non-white pixels
    mask = cv.inRange(image, (0, 0, 0), (200, 200, 200))

    # Find the indices of non-white pixels
    non_white_indices = np.where(mask != 0)

    # Draw circles on non-white pixels
    for y, x in zip(*non_white_indices):
        cv.circle(image, (x, y), 28, (0, 0, 0), 56)

    # Display the resulting image
    cv.imshow('Circles on Non-White Pixels', image)
    cv.waitKey(0)
    cv.destroyAllWindows()

# Example usage
image_path = 'Circle Objects.png'
draw_circle_on_non_white_pixels(image_path)
```

```python
import cv2 as cv
import numpy as np

def draw_circle_on_non_white_pixels(image_path):
    # Load the image
    image = cv.imread(image_path)

    # Create a mask of non-white pixels
    mask = cv.inRange(image, (0, 0, 0), (200, 200, 200))

    # Find the indices of non-white pixels
    non_white_indices = np.transpose(np.nonzero(mask))

    # Draw circles on non-white pixels
    for y, x in non_white_indices:
        cv.circle(image, (x, y), 28, (0, 0, 0), 56)

    # Display the resulting image
    cv.imshow('Circles on Non-White Pixels', image)
    cv.waitKey(0)
    cv.destroyAllWindows()

# Example usage
image_path = 'Circle Objects.png'
draw_circle_on_non_white_pixels(image_path)
```

```python
import cv2
import numpy as np

def calculate_psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    max_pixel = np.max(img1)
    psnr = 10 * np.log10((max_pixel ** 2) / mse)
    return psnr

# โหลดภาพ input และภาพ output
img_input = cv2.imread('input.jpg')
img_output = cv2.imread('output.jpg')

# แปลงภาพให้เป็นแบบ float32
img_input = img_input.astype(np.float32)
img_output = img_output.astype(np.float32)

# คำนวณค่า PSNR
psnr_value = calculate_psnr(img_input, img_output)

print("PSNR value:", psnr_value)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Step 1: Create Sobel Filter (Horizontal) in Spatial Domain
sobel_filter_horizontal = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

# Step 2: Convert Input Image to Frequency Domain
input_image = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)
frequency_domain_input = np.fft.fft2(input_image)

# Display the original image
plt.subplot(2, 3, 1), plt.imshow(input_image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

# Step 3: Convert Sobel Filter to Frequency Domain
frequency_domain_sobel = np.fft.fft2(sobel_filter_horizontal, s=frequency_domain_input.shape)

# Display the Sobel filter (Frequency Domain)
plt.subplot(2, 3, 2), plt.imshow(np.log(1 + np.abs(frequency_domain_sobel)), cmap='gray')
plt.title('Sobel Filter (Frequency Domain)'), plt.xticks([]), plt.yticks([])

# Step 4: Perform Dot Product in Frequency Domain
filtered_image_frequency = frequency_domain_input * frequency_domain_sobel

# Display the result of the dot product (Frequency Domain)
plt.subplot(2, 3, 3), plt.imshow(np.log(1 + np.abs(filtered_image_frequency)), cmap='gray')
plt.title('Result of Dot Product (Frequency Domain)'), plt.xticks([]), plt.yticks([])

# Step 5: Convert Result back to Spatial Domain
filtered_image_spatial = np.fft.ifft2(filtered_image_frequency)
filtered_image_spatial = np.abs(filtered_image_spatial)

# Display the filtered image (Spatial Domain)
plt.subplot(2, 3, 4), plt.imshow(filtered_image_spatial, cmap='gray')
plt.title('Filtered Image (Spatial Domain)'), plt.xticks([]), plt.yticks([])

# Step 6: Compare Results between Spatial and Frequency Domain
filtered_image_spatial_domain = cv2.filter2D(input_image, -1, sobel_filter_horizontal)

# Display the filtered image obtained through spatial filtering
plt.subplot(2, 3, 5), plt.imshow(filtered_image_spatial_domain, cmap='gray')
plt.title('Filtered Image (Spatial Filtering)'), plt.xticks([]), plt.yticks([])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

def enlarge_image(input_image_path, output_image_path, scale_factor=4.0):
    # Open the input image using OpenCV
    input_image = cv2.imread(input_image_path)

    # Enlarge the image using OpenCV with INTER_LANCZOS4 interpolation
    width, height = input_image.shape[1], input_image.shape[0]
    new_width = int(width * scale_factor)
    new_height = int(height * scale_factor)
    resized_image = cv2.resize(
        input_image,
        (new_width, new_height),
        interpolation=cv2.INTER_LANCZOS4  # Change interpolation method for better quality
    )

    # Save the enlarged image
    cv2.imwrite(output_image_path, resized_image)

if __name__ == "__main__":
    # Replace 'input.jpg' with the path to your input image.
    input_image_path = 'IMG_5815.jpg'
    # Replace 'output.jpg' with the desired output image path.
    output_image_path = 'IMG_5815_2.jpg'
    
    # You can adjust the scale_factor to your preference.
    enlarge_image(input_image_path, output_image_path, scale_factor=4.0)
```
